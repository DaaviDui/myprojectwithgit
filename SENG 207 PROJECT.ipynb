```python
import seaborn as sns
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(data[['fixed acidity', 'volatile acidity', 'citric acid', 'alcohol']], figsize=(12, 8))
plt.suptitle('Scatter Plot Matrix of Selected Features')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='quality', y='alcohol', data= data, palette='plasma')
plt.title('Violin Plot of Alcohol Content by Wine Quality')
plt.xlabel('Wine Quality')
plt.ylabel('Alcohol Content')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.kdeplot(x='pH', hue='quality', data= data, fill=True, palette='viridis', common_norm=False)
plt.title('Distribution of pH by Wine Quality')
plt.xlabel('pH')
plt.ylabel('Density')
plt.show()
```

```python
avg_citric_acid = data.groupby('quality')['citric acid'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='quality', y='citric acid', data=avg_citric_acid, palette='viridis')
plt.title('Average Citric Acid Content by Wine Quality')
plt.xlabel('Wine Quality')
plt.ylabel('Average Citric Acid Content')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='alcohol', y='residual sugar', data= data, marker='o', color='coral')
plt.title('Line Plot of Residual Sugar over Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Residual Sugar')
plt.show()
```
